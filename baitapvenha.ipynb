```python
'''1. Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của mảng mới là min của 2 phần tương ứng từ 2 mảng cho trước. Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới.

Ví dụ:

Mảng a: 3 9 1 4

Mảng b: 2 7 4 3 2 8

Mảng kết quả: 2 7 1 3 2 8'''
def tron_mang(a, b):
    c = []
    i = 0

    # Trộn phần chung
    while i < len(a) and i < len(b):
        c.append(min(a[i], b[i]))
        i += 1

    # Thêm phần dư của a
    while i < len(a):
        c.append(a[i])
        i += 1

    # Thêm phần dư của b
    while i < len(b):
        c.append(b[i])
        i += 1

    return c
a = [3, 9, 1, 4]
b=  [2, 7, 4, 3, 2, 8]
print("kết quả là: ",tron_mang(a, b))
```

```python
"""2.	Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
a.	Tính tổng các phần tử thuộc tam giác trên của đường chéo phụ kể cả đường chéo phụ trong ma trận vuông a cấp n.
b.	Chuyển các phần tử âm thành trị tuyệt đối của nó trong a.
c.	Thay các phần tử chẵn trong a bằng số nguyên x cho trước.
d.	Kiểm tra a có toàn chẵn không?
e.	Cho ma trận vuông a cấp n, viết hàm kiểm tra a có đối xứng không. Biết rằng ma trận đối xứng là ma trận có a[i][j] = a[j][i].
f.	Kiểm tra ma trận vuông a cấp n có đường chéo chính tăng dần không?
g.	Xuất các phần tử thuộc tam giác dưới của đường chéo phụ kể cả đường chéo phụ trong ma trận vuông a cấp n.
h.	Kiểm tra ma trận vuông a cấp n có đường chéo phụ giảm dần không?
3.	Tạo một dictionary chứa 3 dictionaries.
"""
import random

# ====== TẠO MA TRẬN RANDOM ======
def tao_ma_tran(n):
    return [[random.randint(-10, 10) for _ in range(n)] for _ in range(n)]

def xuat_ma_tran(a):
    for row in a:
        print(*row)

# ====== a ======
def tong_tam_giac_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j < n - 1:
                tong += a[i][j]
    return tong

# ====== b ======
def doi_am_thanh_duong(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])

# ====== c ======
def thay_chan_bang_x(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x

# ====== d ======
def toan_chan(a):
    for row in a:
        for val in row:
            if val % 2 != 0:
                return False
    return True

# ====== e ======
def doi_xung(a, n):
    for i in range(n):
        for j in range(i+1, n):
            if a[i][j] != a[j][i]:
                return False
    return True

# ====== f ======
def cheo_chinh_tang(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

# ====== g ======
def xuat_tam_giac_duoi_cheo_phu(a, n):
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()

# ====== h ======
def cheo_phu_giam(a, n):
    for i in range(n - 1):
        if a[i][n-1-i] <= a[i+1][n-2-i]:
            return False
    return True

# ====== BÀI 3 ======
def tao_dictionary():
    return {
        "sv1": {"ten": "An", "tuoi": 20, "diem": 8.5},
        "sv2": {"ten": "Binh", "tuoi": 21, "diem": 7.8},
        "sv3": {"ten": "Cuong", "tuoi": 19, "diem": 9.0}
    }

# ====== MAIN ======
def main():
    n = int(input("Nhập n: "))

    a = tao_ma_tran(n)

    print("\nMa trận random:")
    xuat_ma_tran(a)

    print("\n1. Tổng tam giác trên chéo phụ:", tong_tam_giac_tren_cheo_phu(a, n))

    doi_am_thanh_duong(a)
    print("\n2. Sau khi đổi âm -> dương:")
    xuat_ma_tran(a)

    x = int(input("\nNhập x: "))
    thay_chan_bang_x(a, x)
    print("3. Sau khi thay số chẵn:")
    xuat_ma_tran(a)

    print("\n4. Toàn chẵn:", toan_chan(a))
    print("5. Đối xứng:", doi_xung(a, n))
    print("6. Chéo chính tăng:", cheo_chinh_tang(a, n))

    print("\n7. Tam giác dưới chéo phụ:")
    xuat_tam_giac_duoi_cheo_phu(a, n)

    print("\n8. Chéo phụ giảm:", cheo_phu_giam(a, n))

    print("\n9. Dictionary:")
    d = tao_dictionary()
    for sv, info in d.items():
        print(f"{sv} -> {info}")

# ====== RUN ======
main()
```
