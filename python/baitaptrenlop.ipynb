```python
#13.	Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của mảng mới là tổng của 2 phần tương ứng từ 2 mảng cho trước. Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới.
def trong_mang(a, b):
    c = []
    n = min(len(a), len(b))


    for i in range(n):
        c.append(a[i] + b[i])

    for i in range(n, len(a)):
        c.append(a[i])


    for i in range(n, len(b)):
        c.append(b[i])

    return c

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

kq = trong_mang(a, b)
print("Mảng kết quả:", kq)
```

```python
'''14. Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:

a. Tạo mảng a chứa các số nguyên ngẫu nhiên.

b. Xuất các phần tử thuộc dòng k.

c. Xuất các phần tử thuộc cột k.

d. Tìm dòng có tổng lớn nhất nhất 45.

e. Tìm cột có tích nhỏ nhất.

f. Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.

g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.

h. Tính trung bình cộng các phần tử thuộc biên.

i. Tính trung bình tích các phần tử không thuộc biên.'''
import random
import builtins

# a. Tạo ma trận ngẫu nhiên
def tao_ma_tran(m, n):
    return [[random.randint(1, 100) for _ in range(n)] for _ in range(m)]

# Xuất ma trận
def xuat_ma_tran(a):
    for row in a:
        print(row)

# b. Xuất dòng k
def xuat_dong_k(a, k):
    print("Dòng", k, ":", a[k])

# c. Xuất cột k
def xuat_cot_k(a, k):
    print("Cột", k, ":", end=" ")
    for i in range(len(a)):
        print(a[i][k], end=" ")
    print()

# d. Dòng có tổng lớn nhất > 45
def dong_tong_lon_nhat(a):
    max_sum = -1
    index = -1
    for i in range(len(a)):
        s = builtins.sum(a[i])
        if s > max_sum and s > 45:
            max_sum = s
            index = i
    return index, max_sum

# e. Cột có tích nhỏ nhất
def cot_tich_nho_nhat(a):
    m, n = len(a), len(a[0])
    min_product = float('inf')
    index = -1

    for j in range(n):
        prod = 1
        for i in range(m):
            prod *= a[i][j]
        if prod < min_product:
            min_product = prod
            index = j

    return index, min_product

# f. Phần tử dòng chẵn, cột lẻ
def dong_chan_cot_le(a):
    print("Dòng chẵn, cột lẻ:")
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 == 1:
                print(a[i][j], end=" ")
    print()

# g. TBC số chẵn thuộc dòng lẻ
def tbc_chan_dong_le(a):
    tong = dem = 0
    for i in range(len(a)):
        if i % 2 == 1:
            for x in a[i]:
                if x % 2 == 0:
                    tong += x
                    dem += 1
    return tong / dem if dem != 0 else 0

# h. TBC phần tử biên
def tbc_bien(a):
    m, n = len(a), len(a[0])
    tong = dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1

    return tong / dem

# i. TBC phần tử không thuộc biên
def tbc_trong(a):
    m, n = len(a), len(a[0])
    tong = dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            tong += a[i][j]
            dem += 1

    return tong / dem if dem != 0 else 0


# ================= MAIN =================
m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))

a = tao_ma_tran(m, n)

print("\nMa trận:")
xuat_ma_tran(a)

# b, c
xuat_dong_k(a, 1)
xuat_cot_k(a, 2)

# d
print("Dòng có tổng >45 lớn nhất:", dong_tong_lon_nhat(a))

# e
print("Cột có tích nhỏ nhất:", cot_tich_nho_nhat(a))

# f
dong_chan_cot_le(a)

# g
print("TBC chẵn dòng lẻ:", tbc_chan_dong_le(a))

# h
print("TBC biên:", tbc_bien(a))

# i
print("TBC không thuộc biên:", tbc_trong(a))
```

```python
'''15. Một sinh viên gồm các thông tin sau:

Mã sinh viên là chuỗi có 10 ký tự

Tên là chuỗi tối đa 20 ký tự

Năm sinh là một số nguyên

Điểm trung bình là một số thực a.

a. Xây dựng cấu trúc SINHVIEN mô tả một sinh viên

b. Cho một mảng có n sinh viên.

c. Viết hàm cho biết có bao sinh viên đủ điều kiện lên lớp, biết rằng sinh viên đủ điều kiện lên lớp khi điểm trung bình lớn hơn hoặc bằng 5.

d. Xuất các sinh viên đủ 20 tuổi.

e. Đếm số sinh viên học hệ đại học, biết rằng sinh viên hệ DH có mã sinh viên chứa 2 ký tự DH ở vị trí 2,3 trong chuỗi. VD: 02DH0001.

f. Cho biết trong mảng có bao nhiêu sinh viên có tên «Lan»

g. Cho biết trong mảng có bao nhiêu sinh viên có họ «Phan»'''

class SinhVien:
    def __init__(self, masv, hoten, namsinh, dtb):
        self.masv = masv
        self.hoten = hoten
        self.namsinh = namsinh
        self.dtb = dtb


# Nhập danh sách
ds = []
n = int(input("Nhập số sinh viên: "))

for i in range(n):
    print(f"\nSinh viên {i+1}:")
    masv = input("Mã SV: ")
    hoten = input("Họ tên: ")
    namsinh = int(input("Năm sinh: "))
    dtb = float(input("Điểm TB: "))
    ds.append(SinhVien(masv, hoten, namsinh, dtb))


# c. Đếm SV lên lớp
dem_len_lop = 0

# e. Đếm hệ DH
dem_dh = 0

# f. Đếm tên Lan
dem_lan = 0

# g. Đếm họ Phan
dem_phan = 0

print("\nSinh viên đủ 20 tuổi:")

for sv in ds:
    # c
    if sv.dtb >= 5:
        dem_len_lop += 1

    # d
    if 2026 - sv.namsinh >= 20:
        print(sv.masv, "-", sv.hoten)

    # e
    if sv.masv[2:4] == "DH":
        dem_dh += 1

    # f
    if sv.hoten.split()[-1] == "Lan":
        dem_lan += 1

    # g
    if sv.hoten.split()[0] == "Phan":
        dem_phan += 1


# Xuất kết quả
print("\nSố SV lên lớp:", dem_len_lop)
print("Số SV hệ DH:", dem_dh)
print("Số SV tên Lan:", dem_lan)
print("Số SV họ Phan:", dem_phan)
```
