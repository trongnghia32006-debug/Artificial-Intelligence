```python
import math
import random

# Players
PLAYER_X = 1
PLAYER_O = -1

# Function to print the board
def print_board(board):
    symbols = {1: 'X', -1: 'O', 0: ' '}
    for row in board:
        print(" | ".join([symbols[cell] for cell in row]))
        print("---------")

# Function to check for a win
def check_win(board, player):
    # Check rows
    for row in board:
        if all(cell == player for cell in row):
            return True
    # Check columns
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True
    # Check diagonals
    if all(board[i][i] == player for i in range(3)) or \
       all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

# Function to check if the board is full (draw)
def is_board_full(board):
    for row in board:
        if 0 in row:
            return False
    return True

# Function to get all empty cells
def get_empty_cells(board):
    empty_cells = []
    for r in range(3):
        for c in range(3):
            if board[r][c] == 0:
                empty_cells.append((r, c))
    return empty_cells

# Function to check if the game is over (win or draw)
def is_game_over(board):
    return check_win(board, PLAYER_X) or check_win(board, PLAYER_O) or is_board_full(board)

def minimax(board, depth, is_maximizing_player):
    # Base cases:
    # If Player X (maximizing player) wins
    if check_win(board, PLAYER_X):
        return 1
    # If Player O (minimizing player) wins
    if check_win(board, PLAYER_O):
        return -1
    # If it's a draw
    if is_board_full(board):
        return 0

    if is_maximizing_player:
        best_score = -math.inf
        for r, c in get_empty_cells(board):
            board[r][c] = PLAYER_X  # Make the move
            score = minimax(board, depth + 1, False) # Recurse for minimizing player
            board[r][c] = 0         # Undo the move
            best_score = max(best_score, score)
        return best_score
    else:
        best_score = math.inf
        for r, c in get_empty_cells(board):
            board[r][c] = PLAYER_O # Make the move
            score = minimax(board, depth + 1, True) # Recurse for maximizing player
            board[r][c] = 0        # Undo the move
            best_score = min(best_score, score)
        return best_score

# Function to find the best move using Minimax
def find_best_move(board):
    best_score = -math.inf
    best_move = None
    for r, c in get_empty_cells(board):
        board[r][c] = PLAYER_X # Try the move for AI (Player X)
        score = minimax(board, 0, False) # Call minimax for the opponent (Player O)
        board[r][c] = 0        # Undo the move

        if score > best_score:
            best_score = score
            best_move = (r, c)
    return best_move

def play_game_ai_vs_random():
    # Initialize an empty board for each new game
    game_board = [
        [0, 0, 0],
        [0, 0, 0],
        [0, 0, 0]
    ]
    current_player = PLAYER_X # AI starts

    print("\n--- Tic-Tac-Toe (AI vs Random) ---")
    print_board(game_board)

    while not is_game_over(game_board):
        if current_player == PLAYER_X:
            print("AI (X) is thinking...")
            move = find_best_move(game_board)
            if move:
                r, c = move
                game_board[r][c] = PLAYER_X
                print(f"AI (X) played at ({r}, {c})")
            else:
                print("AI could not find a move. This should not happen if game is not over.")
                break
        else:
            print("Random Player (O)'s turn.")
            empty_cells = get_empty_cells(game_board)
            if empty_cells:
                r, c = random.choice(empty_cells)
                game_board[r][c] = PLAYER_O
                print(f"Random Player (O) played at ({r}, {c})")
            else:
                print("No empty cells for Random Player. This should not happen.")
                break

        print_board(game_board)
        current_player *= -1 # Switch player

    if check_win(game_board, PLAYER_X):
        print("AI (X) Wins!")
    elif check_win(game_board, PLAYER_O):
        print("Random Player (O) Wins!")
    else:
        print("It's a Draw!")

# Run the game simulation
play_game_ai_vs_random()
```

```python
import math
import random

# Players
PLAYER_X = 1
PLAYER_O = -1

# Function to print the board
def print_board(board):
    symbols = {1: 'X', -1: 'O', 0: ' '}
    for row in board:
        print(" | ".join([symbols[cell] for cell in row]))
        print("---------")

# Function to check for a win
def check_win(board, player):
    # Check rows
    for row in board:
        if all(cell == player for cell in row):
            return True
    # Check columns
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True
    # Check diagonals
    if all(board[i][i] == player for i in range(3)) or \
       all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

# Function to check if the board is full (draw)
def is_board_full(board):
    for row in board:
        if 0 in row:
            return False
    return True

# Function to get all empty cells
def get_empty_cells(board):
    empty_cells = []
    for r in range(3):
        for c in range(3):
            if board[r][c] == 0:
                empty_cells.append((r, c))
    return empty_cells

# Function to check if the game is over (win or draw)
def is_game_over(board):
    return check_win(board, PLAYER_X) or check_win(board, PLAYER_O) or is_board_full(board)

def minimax_alpha_beta(board, depth, alpha, beta, is_maximizing_player):
    # Base cases:
    if check_win(board, PLAYER_X):
        return 1
    if check_win(board, PLAYER_O):
        return -1
    if is_board_full(board):
        return 0

    if is_maximizing_player:
        best_score = -math.inf
        for r, c in get_empty_cells(board):
            board[r][c] = PLAYER_X
            score = minimax_alpha_beta(board, depth + 1, alpha, beta, False)
            board[r][c] = 0
            best_score = max(best_score, score)
            alpha = max(alpha, best_score)
            if beta <= alpha:
                break  # Alpha-beta pruning
        return best_score
    else:
        best_score = math.inf
        for r, c in get_empty_cells(board):
            board[r][c] = PLAYER_O
            score = minimax_alpha_beta(board, depth + 1, alpha, beta, True)
            board[r][c] = 0
            best_score = min(best_score, score)
            beta = min(beta, best_score)
            if beta <= alpha:
                break  # Alpha-beta pruning
        return best_score

# Function to find the best move using Alpha-Beta Pruning
def find_best_move_alpha_beta(board):
    best_score = -math.inf
    best_move = None
    alpha = -math.inf
    beta = math.inf

    for r, c in get_empty_cells(board):
        board[r][c] = PLAYER_X
        score = minimax_alpha_beta(board, 0, alpha, beta, False) # Call minimax for the opponent (Player O)
        board[r][c] = 0

        if score > best_score:
            best_score = score
            best_move = (r, c)
        alpha = max(alpha, best_score)

    return best_move

def play_game_ai_vs_random_alpha_beta():
    game_board = [
        [0, 0, 0],
        [0, 0, 0],
        [0, 0, 0]
    ]
    current_player = PLAYER_X # AI starts

    print("\n--- Tic-Tac-Toe (AI with Alpha-Beta vs Random) ---")
    print_board(game_board)

    while not is_game_over(game_board):
        if current_player == PLAYER_X:
            print("AI (X) is thinking (Alpha-Beta)...")
            move = find_best_move_alpha_beta(game_board)
            if move:
                r, c = move
                game_board[r][c] = PLAYER_X
                print(f"AI (X) played at ({r}, {c})")
            else:
                print("AI could not find a move. This should not happen if game is not over.")
                break
        else:
            print("Random Player (O)'s turn.")
            empty_cells = get_empty_cells(game_board)
            if empty_cells:
                r, c = random.choice(empty_cells)
                game_board[r][c] = PLAYER_O
                print(f"Random Player (O) played at ({r}, {c})")
            else:
                print("No empty cells for Random Player. This should not happen.")
                break

        print_board(game_board)
        current_player *= -1 # Switch player

    if check_win(game_board, PLAYER_X):
        print("AI (X) Wins!")
    elif check_win(game_board, PLAYER_O):
        print("Random Player (O) Wins!")
    else:
        print("It's a Draw!")

# Run the game simulation
play_game_ai_vs_random_alpha_beta()
```
