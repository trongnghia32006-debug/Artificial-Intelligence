```python
import math

X = "X"
O = "O"
EMPTY = None
```

```python
def initial_state():
    return [[EMPTY for _ in range(3)] for _ in range(3)]


def actions(board):
    return [(i, j) for i in range(3) for j in range(3)
            if board[i][j] is EMPTY]


def result(board, action):
    i, j = action
    new_board = [row.copy() for row in board]
    new_board[i][j] = X
    return new_board


def terminal(board):
    return False


def utility(board):
    return 0
```

```python
def max_value_ab(board, alpha, beta):

    v = -math.inf

    for a in actions(board):

        v = max(v,
                min_value_ab(
                    result(board, a),
                    alpha,
                    beta
                ))

        if v >= beta:
            return v

        alpha = max(alpha, v)

    return v


def min_value_ab(board, alpha, beta):

    v = math.inf

    for a in actions(board):

        v = min(v,
                max_value_ab(
                    result(board, a),
                    alpha,
                    beta
                ))

        if v <= alpha:
            return v

        beta = min(beta, v)

    return v
```

```python
def alphabeta(board):

    best = -math.inf
    move = None

    for a in actions(board):

        val = min_value_ab(
            result(board, a),
            -math.inf,
            math.inf
        )

        if val > best:
            best = val
            move = a

    return move
```

```python
board = initial_state()
print("Bai 2 - AlphaBeta move:", alphabeta(board))

```
