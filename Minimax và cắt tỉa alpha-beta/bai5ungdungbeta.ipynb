```python
import math
```

```python
class GameState:

    def actions(self):
        pass

    def result(self, action):
        pass

    def terminal(self):
        pass

    def utility(self):
        pass
```

```python
def alphabeta_generic(state, alpha, beta, maximizing):

    if state.terminal():
        return state.utility()

    if maximizing:
        value = -math.inf

        for a in state.actions():
            value = max(value,
                        alphabeta_generic(
                            state.result(a),
                            alpha,
                            beta,
                            False
                        ))

            if value >= beta:
                break

            alpha = max(alpha, value)

        return value

    else:
        value = math.inf

        for a in state.actions():
            value = min(value,
                        alphabeta_generic(
                            state.result(a),
                            alpha,
                            beta,
                            True
                        ))

            if value <= alpha:
                break

            beta = min(beta, value)

        return value
```
